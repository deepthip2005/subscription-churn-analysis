```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
# Make sure you have uploaded 'Customer-Churn.csv' to the folder on the left!
df = pd.read_csv('Customer-Churn.csv')

# 2. Data Cleaning
# Converting 'TotalCharges' to numbers because it often loads as text
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna()

# 3. Visualization: Monthly Charges vs Churn
plt.figure(figsize=(10, 5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette='magma')
plt.title('Monthly Charges Impact on Churn')
plt.show()

# 4. Visualization: Contract Type vs Churn
plt.figure(figsize=(10, 5))
sns.countplot(x='Contract', hue='Churn', data=df, palette='viridis')
plt.title('Churn Count by Contract Type')
plt.show()
```
